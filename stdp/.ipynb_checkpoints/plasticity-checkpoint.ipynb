```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(69)
```

### Exercise 1: Bienenstock Cooper Munro (BCM) Rule

```python
def bcm():
  xs = np.array([[20, 0], [0, 20]])
  p = 0.5

  dt = 1  # ms
  T = int(10e3 / dt)  # ms
  eta = 1e-6  # ms^-1
  tau = 50  # ms
  y0 = 10

  y = np.zeros((T, 1))
  theta = np.zeros((T, 1))
  w = np.zeros((T, 2))
  w[0] = [0.5, 0.5]
  theta[0] = 15
  x1_mask = np.random.choice([True, False], size=T, p=[p, 1-p])

  for t in range(T-1):
    x = xs[:,int(~x1_mask[t])]
    y[t] = w[t][None,:] @ x
    theta[t+1] = theta[t] + dt / tau * (-theta[t] + y[t]**2 / y0)
    w[t+1] = w[t] + eta * x * y[t] * (y[t] - theta[t])
    w[t+1] = np.maximum(w[t+1], 0)

  return w, theta.squeeze(), y.squeeze(), x1_mask
```

```python
w, theta, y, x1_mask = bcm()

def mask(a, mask):
    a = a * mask
    a[0] = a[mask][0]
    for i in range(len(a)-1):
        a[i+1] = a[i] * ~(mask[i+1]) + a[i+1]
    return a

def plot_window(start=0, end=-1, theta_alpha=0.4, y_lim_top=None, 
                time_formatter=lambda x, pos: f"{x/1e3:.0f}s", time_unit="s"):
    fig, (a1, a2) = plt.subplots(2, 1, figsize=(10,5), tight_layout=True, sharex=True)
    
    a1.grid()
    a1.plot(y[start:end], 'k', alpha=0.2, lw=0.8)
    a1.plot(theta[start:end], 'g', lw=1, alpha=theta_alpha, label= r"$\theta$")
    a1.plot(mask(y, ~x1_mask)[start:end], 'k--', label="$y | x_2$")
    a1.plot(mask(y, x1_mask)[start:end], 'k:', label="$y | x_1$")
    a1.set_ylim(top=y_lim_top)
    a1.set_ylabel("Post-neuron")
    a1.legend()
    
    a2.grid()
    a2.plot(w[:, 1][start:end], 'k--')
    a2.plot(w[:, 0][start:end], 'k:')
    a2.set_ylabel("Synaptic weights")
    a2.legend(["$w_2$", "$w_1$"])
    a2.xaxis.set_major_formatter(time_formatter)
    
    fig.supxlabel(f"Time ({time_unit})")
    fig.suptitle(" Bienenstock Cooper Munro Rule")

    # plt.savefig("bcm.svg")
    plt.show()

plot_window()
plot_window(end=1000, theta_alpha=0.7, y_lim_top=y[:1000].max() + 0.1, 
            time_formatter=lambda x, pos: f"{x:.0f}ms", time_unit="ms")
```

### Exercise 2: Spike-Timing Dependent Plasticity (STDP)

```python
def stdp():
    A_p = A_m = 1
    tau_p, tau_m = 10, 20 # ms
    dt = 1 # ms
    spike_dt = 1000 # ms
    num_pairs = 60
    lags = np.arange(-50,51,3) # ms
    dw = np.zeros(len(lags))
    spikes = np.tile(np.array([1] + [0] * (spike_dt - 1)), num_pairs)

    for i, lag in enumerate(lags):
        pre_spikes = spikes.copy()
        post_spikes = np.roll(spikes, abs(lag))
        if lag > 0: pre_spikes, post_spikes = post_spikes, pre_spikes

        x, y = 0, 0
        for t in range(len(pre_spikes)):
            x  += dt / tau_p * (-x + pre_spikes[t])
            y  += dt / tau_m * (-y + post_spikes[t])
            dw[i] += A_p * x * post_spikes[t] - A_m * y * pre_spikes[t]

    return lags, dw
```

```python
lags, dw = stdp()

plt.figure(figsize=(6, 5), tight_layout=True) 
plt.grid()
plt.plot(lags[lags < 0], dw[lags < 0], 'k', lw=2)
plt.plot(lags[lags > 0], dw[lags > 0], 'k', lw=2)
plt.title(r"Spike-Timing Dependent Plasticity")
plt.ylabel(r"$\Delta w$", rotation=0, fontsize=12)
plt.xlabel(r"$t^{pre} - t^{post}$ (ms)")
plt.savefig("stdp.svg")
plt.show()
```
